```python
from sympy.plotting import plot
from sympy import Symbol, solve, diff, evalf, solveset, S, pprint, sin, cos 
```

```python
x = Symbol('x')
```

```python
# 1. Построить график

f = sin(x)**2 - cos(x)**2
p1 = plot(f, show=False)
p1.show()
```

```python
# 2. Определить корни

ans = solve(f,x)
print(ans)

for i in ans:
    print(i.evalf())
```

```python
# 3. Определить промежутки, на котором f > 0
# Вариант 1

pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# Вариант 2

rande_plus = solve(f > 0,x)
rande_plus
```

```python
# 4. Определить промежутки, на котором f < 0
# Вариант 1

pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
# Вариант 2

range_minus = solve(f < 0,x)
range_minus
```

```python
# 5. Вычислить вершину

diff_f = diff(f,x)
print(diff_f)
solve(diff_f,x)
```

```python
# 6. Найти интервалы, на которых функция возрастает

pprint(solveset(diff_f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 7. Найти интервалы, на которых функция убывает

pprint(solveset(diff_f < 0, x, domain=S.Reals), use_unicode=True)
```
